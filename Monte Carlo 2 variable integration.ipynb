```python
import random
import math
import statistics
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Monte_Carlo:

    def __init__(self, lower_boundx, upper_boundx, lower_boundy, upper_boundy):
        self.lower_boundx = lower_boundx
        self.upper_boundx = upper_boundx
        self.lower_boundy = lower_boundy
        self.upper_boundy = upper_boundy
        
        self.random_listx = []
        self.random_listy = []
        self.function_values = []
        
    def function(self, x, y):
        return x * y * np.exp(-((x - 1) ** 2))
        
    def integral(self):
        samples = 1000000

        for i in range(samples):
            n = random.uniform(self.lower_boundx, self.upper_boundx)
            self.random_listx.append(n)

        for i in range(samples):
            n = random.uniform(self.lower_boundy, self.upper_boundy)
            self.random_listy.append(n)

        sum = 0
        for q in range(len(self.random_listx)):
            self.function_values.append(self.function(self.random_listx[q], self.random_listy[q]))
            sum += self.function_values[q]

        integral_value = sum * (self.upper_boundx - self.lower_boundx) * (self.upper_boundy - self.lower_boundy) / samples
        print(integral_value)
        
    def graph(self):
        print('3D graphs are loading')
        x = np.array(self.random_listx)
        y = np.array(self.random_listy)
        z = np.array(self.function_values)

        ax = plt.axes(projection='3d')
        ax.scatter(x, y, z, c=z, cmap='prism_r', linewidth=0.1)
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        ax.set_title('sampled points');

        x = np.linspace(self.lower_boundx, self.upper_boundx, 10)
        y = np.linspace(self.lower_boundy, self.upper_boundy, 10)

        X, Y = np.meshgrid(x, y)
        Z = self.function(X, Y)
        fig = plt.figure()
        ax = plt.axes(projection='3d')
        ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='prism_r', edgecolor='red')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        ax.set_title('function');
    

lower_boundx = -2
upper_boundx = 3
lower_boundy = 0
upper_boundy = 1

mc = Monte_Carlo(lower_boundx, upper_boundx, lower_boundy, upper_boundy)
mc.integral()
mc.graph()
```
