```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = float(input("Enter the launch angle in degrees "))

#converts from degrees to radians
theta = theta * math.pi/(180)

vi = float(input('Enter the initial velocity at that angle '))

#inital x and y velocity
viy = vi * math.sin(theta)
vix = vi * math.cos(theta)
acc = -9.81

#initial and final velocities are equal
viyf = -viy

#solving for time
time = (viyf - viy)/acc
time = int(100*round(time,1))
#print('time= ',time)

#horizontal distance
dx = []
for t in range(time):
    dx.append(vix*(t/100))
#print(dx)
    
#vertical distance
dy = []
for t in range(time):
    dy.append(viy*t/100+acc/2*t/100*t/100)
#print(dy)

plt.plot(dx,dy)
```

```python
#check for errors
theta = float(input("Enter the launch angle in degrees "))

mass = 100

#drag force coefficient and object area
#fluid density = 1
drag = 0.5
area = 10

#converts from degrees to radians
theta = theta * math.pi/(180)

vi = float(input('Enter the initial velocity at that angle '))

#inital x and y velocity
acc = -9.81
vt = np.sqrt(2*mass*-acc/(drag*area))
time = vt/(-acc)*math.atan(vi/vt)
viy = vi * math.sin(theta)*np.exp(-acc*(time)/(vt))-vt*(1-np.exp(-acc*(time)/(vt)))
vix = vi * math.cos(theta)*np.exp(-acc*(time)/(vt))

time = int(100*round(time,1))

#horizontal distance
dx = []
for t in range(time):
    dx.append(vix*vt/acc*(1-np.exp(-acc*(t/100)/(vt))))
#print(dx)
    
#vertical distance
dy = []
for t in range(time):
    dy.append(vt/acc*(viy+vt)*(1-np.exp(-acc*(t/100)/(vt)))-vt*t/100)
#print(dy)

plt.plot(dx,dy)
```
